```python

import pandas as pandas

#loading the dataset
df = pandas.read_csv("C:\\Users\\kayce\\Downloads\\StudentsPerformance.csv")
df.head() #checking the first 5 rows of dataset
```

```python
df.info() #Overview of the columns and data types
df.describe() #Statistical summary of numerical columns
```

```python
math_mean = df.groupby('test preparation course')['math score'].mean() 
print(math_mean, "\n")# Average math score based on test preparation course completion
reading_mean = df.groupby('test preparation course')['reading score'].mean() 
print(reading_mean, "\n")# Average reading score based on test preparation course completion
writing_mean = df.groupby('test preparation course')['writing score'].mean() 
print(writing_mean, "\n")# Average writing score based on test preparation course completion
```

```python
correlation = df[['math score', 'reading score', 'writing score']].corr()
print(correlation) #Correlation matrix for the scores
```

```python
import matplotlib.pyplot as plt
df.boxplot(column=['math score'], by='test preparation course')
df.boxplot(column=['reading score'], by='test preparation course')
df.boxplot(column=['writing score'], by='test preparation course')  
```
